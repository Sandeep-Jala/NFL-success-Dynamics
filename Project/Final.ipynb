### How do the datasets complement each other?

- **NFL Statistics Dataset**: This dataset includes game-by-game statistics such as scores, team performances, win/loss records, and player stats. It provides the foundational data for understanding the outcomes of games and the performance metrics of teams and individual players.

- **Weather Dataset**: Contains detailed information about weather conditions for each game, including temperature, wind speed, humidity, and specific weather events (rain, snow, etc.). Weather conditions can significantly affect game strategies and player performance, especially in outdoor stadiums, making this dataset crucial for analyzing game outcomes in relation to environmental factors.

- **Betting Odds Dataset**: Offers insights into how games are perceived by the public and the betting markets before they happen, including the favored team and predicted point spreads and over/under lines. This dataset helps gauge public expectations and market sentiment, which can be compared against actual game outcomes and weather conditions to identify patterns or biases.

- **Quarterback Performance Dataset**: Focuses on the performance metrics of starting quarterbacks, such as touchdowns, completions, and passing yards. Quarterbacks play a pivotal role in their teams' offensive strategies, making their performance data essential for analyzing its impact on game outcomes, especially when cross-referenced with weather conditions and betting odds.

- **Crime Data Dataset**: There are 2 datasets, one on Robbery and one on Aggravated Assault. These datasets offer a unique perspective on the socio-economic environment surrounding the teams, potentially correlating with team performance, fan engagement, and even betting odds. It allows for the exploration of how external societal factors might influence or reflect the dynamics of NFL games.

## Preprocess NFL_QB_STATS

```python
nfl_qb_stats['G'] = pd.to_numeric(nfl_qb_stats['G'])
nfl_qb_stats_sorted = nfl_qb_stats.sort_values(['Team', 'G'], ascending=[True, False])
nfl_qb_stats_cleaned = nfl_qb_stats_sorted.drop_duplicates(subset=['Team'], keep='first')
nfl_qb_stats_cleaned.reset_index(drop=True, inplace=True)
nfl_qb_stats_cleaned
```

Processed the NFL quarterback dataset, focusing on highlighting the most active quarterback for each team. Initially, the 'G' column, which is number of games played, is converted to a numeric data type to facilitate sorting. Following the sorting, duplicates are removed, keeping only the entry with the highest number of games played for each team, effectively selecting the most active or starting quarterback.

```python
team_to_qb_dict = nfl_qb_stats_cleaned.set_index('Team')['Player'].to_dict()
team_to_qb_dict
```

```python
# Create mappings for the requested QB stats
# We will sum passing TDs and rushing TDs for the total TDs per QB
nfl_qb_stats_cleaned['Total_TD'] = nfl_qb_stats_cleaned['TD'] + nfl_qb_stats_cleaned['TD.1']

# Mapping dictionaries for the stats
team_to_td_dict = nfl_qb_stats_cleaned.set_index('Team')['Total_TD'].to_dict()
team_to_comp_dict = nfl_qb_stats_cleaned.set_index('Team')['Comp'].to_dict()
team_to_yards_dict = nfl_qb_stats_cleaned.set_index('Team')['Yard'].to_dict()
team_to_games_dict = nfl_qb_stats_cleaned.set_index('Team')['G'].to_dict()

# Convert 'Yard' strings to numeric values (removing commas)
nfl_qb_stats_cleaned['Yard'] = nfl_qb_stats_cleaned['Yard'].str.replace(',', '').astype(int)
team_to_yards_dict = nfl_qb_stats_cleaned.set_index('Team')['Yard'].to_dict()

# Map the QB stats to the home and away teams in the nfl_betting_weather_crime dataset
merged_data['TD_QB_home'] = merged_data['team_home'].map(team_to_td_dict)
merged_data['TD_QB_away'] = merged_data['team_away'].map(team_to_td_dict)
merged_data['Comp_QB_home'] = merged_data['team_home'].map(team_to_comp_dict)
merged_data['Comp_QB_away'] = merged_data['team_away'].map(team_to_comp_dict)
merged_data['Yards_QB_home'] = merged_data['team_home'].map(team_to_yards_dict)
merged_data['Yards_QB_away'] = merged_data['team_away'].map(team_to_yards_dict)
merged_data['Games_QB_home'] = merged_data['team_home'].map(team_to_games_dict)
merged_data['Games_QB_away'] = merged_data['team_away'].map(team_to_games_dict)

# Check the first few rows to ensure the stats were mapped correctly
merged_data[['team_home', 'TD_QB_home', 'Comp_QB_home', 'Yards_QB_home', 'team_away', 'TD_QB_away', 'Comp_QB_away', 'Yards_QB_away','Games_QB_home','Games_QB_away']].head()
```

```python
# Aggregating home and away wins for each state
home_wins_by_state = merged_data.groupby('state')['home_win'].sum().reset_index()
away_wins_by_state = merged_data.groupby('state')['away_win'].sum().reset_index()

# Combining home and away wins by state
wins_by_state = pd.merge(home_wins_by_state, away_wins_by_state, on='state', how='outer')
wins_by_state['total_wins'] = wins_by_state['home_win'] + wins_by_state['away_win']

avg_crime_by_state = merged_data.groupby('state')[['Total Aggravated Assaults','Population']].mean().reset_index()

# Merging win data with crime data
state_analysis = pd.merge(wins_by_state, avg_crime_by_state, on='state')
```

```python
state_analysis['crime_rate']=state_analysis['Total Aggravated Assaults']/state_analysis['Population']
```

```python
# Visualizing the relationship between total wins and crime rates by state
plt.figure(figsize=(14, 8))
sns.scatterplot(x='total_wins', y='crime_rate', hue='state', data=state_analysis, palette='Set2', s=150)
plt.title('Total Wins vs Crime Rates by State')
plt.xlabel('Total Wins (Home and Away)')
plt.ylabel('Average Crime Rate')
plt.legend(title='State', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

In this plot the x-axis represents the total number of wins, and the y-axis represents the average crime rate.
